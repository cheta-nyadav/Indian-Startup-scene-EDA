### using Automation for file handling
<i>Since, the prevoius way seems too manual, we use automation to read the csv files from respective folers. This also improves scalability</i>

```python
import os
folder_path = r"C:\Users\lenovo\Downloads\funding 2022"

# Get all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Read and store all dataframes in a list
dataframes = []
for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path, encoding='ISO-8859-1')  # Use appropriate encoding
    dataframes.append(df)
    print(f"Loaded: {file}")
    
fund2021=pd.read_csv(r"C:\\Users\\lenovo\\Downloads\\funding 2021\\Indian startups funding in 2021.csv")

# Combine all DataFrames into one (if needed)
fund2022=pd.concat(dataframes, ignore_index=True)
```

### Scaling values less than a 10,000 to right scale
<i> as they occure from human/ data entry errors

```python
def scale_amount(val):
    try:
        float_val = float(val)
        if float_val < 10000.0:
            return float_val*(10**6)
        else:
            return(float_val)
    except ValueError:
        return val
        

fund['amount'] = fund['amount'].apply(scale_amount)
```
